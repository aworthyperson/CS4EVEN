```python
UWASH= pd.read_csv("UWASH.2023-02-06.csv", sep=',', header=[10])
UWASH.head()
```

```python
#UWASH= UWASH.drop(['Station_ID'], axis=1)

#drop units 
UWASH= UWASH.drop(0,axis=0)

UWASH.Date_Time= pd.to_datetime(UWASH.Date_Time)
UWASH['Date_Time']= UWASH['Date_Time']

#convert all to numbers besides the wind, station, date
cols = UWASH.columns.drop(['Date_Time', "Station_ID"]) #,'wind_cardinal_direction_set_1d'])
#cols = UWASH.columns.drop(['Date_Time'])

UWASH[cols] = UWASH[cols].apply(pd.to_numeric, errors='coerce')

UWASH= UWASH.dropna(axis=1, how='all')
UWASH.head()
#UWASH.info()
```

```python
KSEA= pd.read_csv("KSEA.2023-02-06.csv", sep=',', header=[10])
KSEA.head()
```

```python
#UWASH= UWASH.drop(['Station_ID'], axis=1)

#drop units 
KSEA= KSEA.drop(0,axis=0)

KSEA.Date_Time= pd.to_datetime(KSEA.Date_Time)
KSEA['Date_Time']= KSEA['Date_Time']

#convert all to numbers besides the wind, station, date
cols = KSEA.columns.drop(['Date_Time', "Station_ID"]) #,'wind_cardinal_direction_set_1d'])


KSEA[cols] = KSEA[cols].apply(pd.to_numeric, errors='coerce')

KSEA= KSEA.dropna(axis=1, how='all')
KSEA.head()
```

```python
print(UWASH.air_temp_set_1.mean())
print(KSEA.air_temp_set_1.mean())
```

## Group Datasets by hourly mean and then compare observations

```python
UWASH.set_index('Date_Time')
UWASH_hr= UWASH.groupby(pd.Grouper(key='Date_Time', axis=0, freq='H')).air_temp_set_1.mean()
UWASH_hr=UWASH_hr.to_frame()
UWASH_hr= UWASH_hr.rename(columns= {"air_temp_set_1": "UWASH_Temperature"})
UWASH_hr.head()
```

```python
KSEA.set_index('Date_Time')
#KSEA_hr= KSEA.groupby([KSEA['Date_Time'].dt.date.hour, KSEA['Date_Time'].dt.hour]).air_temp_set_1.mean()
KSEA_hr= KSEA.groupby(pd.Grouper(key='Date_Time', axis=0, freq='H')).air_temp_set_1.mean()
KSEA_hr=KSEA_hr.to_frame()
KSEA_hr= KSEA_hr.rename(columns= {"air_temp_set_1": "KSEA_Temperature"})
KSEA_hr.head()
```

```python
hr_df= KSEA_hr.join(UWASH_hr)
hr_df.reset_index()

hr_df['dif']= (hr_df.KSEA_Temperature - hr_df.UWASH_Temperature).abs()
hr_df['max_obs']= hr_df.idxmax(axis=1)
hr_df
```

```python
nan_count = hr_df.isna().sum()
print(nan_count )
```

```python
print(hr_df.max_obs.value_counts())
```

```python
x1 = hr_df['KSEA_Temperature']
x2 = hr_df['UWASH_Temperature']

fig, axs1 = plt.subplots(figsize=(11,7))

colors = ['slategrey', 'lightgreen']
names = ['KSEA Temperature', 'UWASH Temperature']

# Make the histogram using a list of lists
axs1.hist([x1,x2], bins = 20, color = colors, label=names)

# Plot formatting
axs1.legend()
axs1.set_xlabel('Outdoor Temperatures (F)')
axs1.set_ylabel('Number of Observations')
axs1.set_title('Side-by-Side Histogram of Observed Outdoor Temperatures at two different stations');
```
